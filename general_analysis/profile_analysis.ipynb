# Profile Data Analysis

The purpose of this workbook is to analyse the data scraped from MP's twitter profiles. Some questions ask?

* Who is the most popular mp on twitter?
* Who tweets the most?
* Who registered first?
* Who has the most followers?
* Which parties have the most/least followers?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

## Profile Data Analysis

The file profile_data.csv contains data scraped from MP's twitter profiles as well as some data from the 2017 election e.g. number of votes.

Note that numbers such as tweet_count may not match up with the csv containing the actual tweets. This is because of:

1. Slight time difference in time collection. Tweets were collected a day or so after the profile data.
2. Some metrics such as tweet count include retweets which were not collected.
3. Not all tweets could be collected due to errors in the search API.


```python
profile_df = pd.read_csv('profile_data.csv', index_col=False)
```

```python
profile_df.head()
```

```python
profile_df.describe()
```

```python
# Merge Labour and Labour and Co-operative

def merge_labour(row):
    
    if row['party'] == 'Labour and Co-operative':
        
        party = 'Labour'
        
    elif row['party'] == 'Scottish National Party':
        
        party = 'SNP'
        
    elif row['party'] == 'Democratic Unionist Party':
        
        party = 'DUP'
        
    else:
        
        party = row['party']
        
    return party

profile_df['party'] = profile_df.apply(merge_labour, axis=1)
```

## MP Rankings

Which MPs have the most followers, tweets, likes etc?

```python
# Most followers

profile_df.sort_values('follower_count', ascending=False).head()
```

```python
# Visualise

fig, ax = plt.subplots(figsize=(6,9))

sns.set(style="whitegrid")
temp =profile_df.sort_values('follower_count', ascending=False).head(30)

sns.stripplot(x='follower_count', y='name', data=temp,size=5, orient="h",palette="Reds_r", edgecolor="gray")

sns.despine(left=True, bottom=True)

ax.xaxis.grid(False)
ax.yaxis.grid(True)
ax.set(xlim=(0, 2000000), xlabel=" Follower Count", ylabel="Name")
plt.title('Which MPs Have the Most Followers')
```

```python
# Repeat for tweet count

fig, ax = plt.subplots(figsize=(6,9))

sns.set(style="whitegrid")
temp =profile_df.sort_values('tweet_count', ascending=False).head(30)

sns.stripplot(x='tweet_count', y='name', data=temp,size=5, orient="h",palette="Reds_r", edgecolor="gray")

sns.despine(left=True, bottom=True)

ax.xaxis.grid(False)
ax.yaxis.grid(True)
ax.set(xlim=(0,90000),xlabel="Tweet Count", ylabel="Name")
plt.title('Which MPs Tweet the Most')
```

## Per Party Analysis

Analyse the data on a per party basis.

```python
order = list(profile_df[['follower_count', 'party']].groupby('party').mean().sort_values('follower_count', ascending=True).index)


sns.barplot(x='party', y='follower_count', data=profile_df, order=order)
```

```python
# Followers

sns.stripplot(x='party', y='follower_count', data=profile_df, jitter=True)
plt.xticks(rotation=45)
```

```python
#Can't see too much on the above plot. Let's try log(follower_count)

profile_df['log_follower_count'] = profile_df.apply(lambda x: np.log10(x['follower_count']), axis=1)
```

```python
order = list(profile_df[['log_follower_count', 'party']].groupby('party').mean().sort_values('log_follower_count').index)

sns.stripplot(x='party', y='log_follower_count', data=profile_df, order=order, jitter=True, palette='Set2')
plt.xticks(rotation=45)
plt.title('Follower Count by Party')
```

```python
#Joined Date
def date_delta(row):
    """
    Number of days before now() that the user joined twitter
    """
    
    dd = datetime.now()-row['joined']
    
    return dd.days


profile_df['time_delta'] = profile_df.apply(date_delta, axis=1)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

order = list(profile_df[['time_delta', 'party']].groupby('party').mean().sort_values('time_delta', ascending=True).index)

sns.stripplot(x='party', y='joined', data=profile_df, order=order, jitter=True)
plt.title('Join Date by Party')
plt.ylabel('Join Date')
plt.xlabel('Party')
```

```python
#Tweet Count

fig, ax = plt.subplots(figsize=(10,5))

order = list(profile_df[['tweet_count', 'party']].groupby('party').mean().sort_values('tweet_count', ascending=True).index)

sns.barplot(x='party', y='tweet_count', data=profile_df, order=order)
plt.title('Mean Tweet Count by Party')
plt.ylabel('Mean Tweet Count')
plt.xlabel('Party')
```

```python
#Photo Count

fig, ax = plt.subplots(figsize=(10,5))

order = list(profile_df[['photo_count', 'party']].groupby('party').mean().sort_values('photo_count', ascending=True).index)

sns.violinplot(x='party', y='photo_count', data=profile_df, order=order, inner='quartiles', palette="Set2", scale="width")
plt.title('Photo Count by Party')
plt.ylabel('Photo Count')
plt.xlabel('Party')
```
